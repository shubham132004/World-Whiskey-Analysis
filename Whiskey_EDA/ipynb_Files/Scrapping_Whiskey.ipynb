```python
varieties = ['https://www.htfw.com/world-whisky/american-whiskey/bourbon-whiskey',
             'https://www.htfw.com/world-whisky/american-whiskey/rye-whiskey',
             'https://www.htfw.com/world-whisky/american-whiskey/tennessee-whiskey',
             'https://www.htfw.com/world-whisky/american-whiskey/kentucky-whiskey',
             'https://www.htfw.com/world-whisky/american-whiskey/buffalo-trace',
             'https://www.htfw.com/brands/bushmills',
             'https://www.htfw.com/brands/midleton',
             'https://www.htfw.com/brands/redbreast',
             'https://www.htfw.com/world-whisky/irish-whiskey/spot',
             'https://www.htfw.com/brands/jameson',
             'https://www.htfw.com/brands/writers_tears',
             'https://www.htfw.com/world-whisky/japanese-whisky',
             'https://www.htfw.com/world-whisky/indian-whisky',
             'https://www.htfw.com/world-whisky/english-whisky',
             'https://www.htfw.com/world-whisky/canadian-whisky',
             'https://www.htfw.com/world-whisky/welsh-whisky',
             'https://www.htfw.com/world-whisky/swedish-whisky',
             'https://www.htfw.com/brands/jack_daniel_s',
             'https://www.htfw.com/brands/yamazaki',
             'https://www.htfw.com/brands/hibiki',
             'https://www.htfw.com/brands/woodford_reserve',
             'https://www.htfw.com/world-whisky/american-whiskey/buffalo-trace',
             'https://www.htfw.com/brands/midleton'
             'https://www.htfw.com/world-whisky/american-whiskey',
             'https://www.htfw.com/world-whisky/irish-whiskey',
             'https://www.htfw.com/world-whisky',
             'https://www.htfw.com/brands/starward',
             'https://www.htfw.com/brands/shizuoka',
             'https://www.htfw.com/brands/waterford',
             'https://www.htfw.com/brands/teeling']

len(varieties)
```

```python
BRAND = []
YEAR = []
BRAND_VARIATION = []
WHISKEY_TYPE = []
CENTI_LITER = []
ABV = []
PRICE = []
COUNTRY = []

for variety in varieties:

    # url = f'https://www.htfw.com/world-whisky?country={countries}'
    url = f'{variety}'

    page = requests.get(url)

    soup = BeautifulSoup(page.text)

    section = soup.find_all('div' , class_ = 'product-info flex grow flex-col gap-2')

    
    for i in section:
        #Brand Name
        brand = i.find('span' , class_ = 'font-serif font-bold line-clamp-1')
        if brand:
            BRAND.append(brand.text.strip())
        else:
            BRAND.append(np.nan)
    
        #Year - How many years old
        year = i.find('span' , class_ = 'flex gap-x-1 text-nowrap flex-wrap font-semibold overflow-hidden text-sm h-5')
        # year = re.findall(r'(\d+ \w+ \w+|\d+ \w+)' , y)
        if year:
            YEAR.append(year.text.strip())
        else:
            YEAR.append(np.nan)
    
        #Brand Variation
        b = i.find('span' , class_ = 'line-clamp-3 text-sm')
        brand_variation = re.findall(r'(.*) \w+' , b.text) #(.*) \w+
        if brand_variation:
            BRAND_VARIATION.append(brand_variation[0].strip())
        else:
            BRAND_VARIATION.append(np.nan)
    
        #Whisky Type
        wt = i.find('span' , class_ = 'line-clamp-3 text-sm')
        whiskey_type = re.findall(r'\w+ Whisky|\w+ \w+' , wt.text)
        if whiskey_type:
            WHISKEY_TYPE.append(whiskey_type[0].strip())
        else:
            WHISKEY_TYPE.append(np.nan)
    
        #cl - Centi-Liters
        cl = i.find('div' , class_ = 'flex grow gap-2 items-end font-semibold text-neutral-600 text-xs')
        centi_liter = re.findall(r'\d+cl' , cl.text)
        if centi_liter:
            CENTI_LITER.append(centi_liter[0].strip())
        else:
            CENTI_LITER.append(np.nan)
    
        #ABV - Alcohol By Volumn
        t2 = i.find('div' , class_ = 'flex grow gap-2 items-end font-semibold text-neutral-600 text-xs')
        abv = re.findall(r'\w+% ABV' , t2.text)
        if abv:
            ABV.append(abv[0].strip())
        else:
            ABV.append(np.nan)
    
        #Price in EURO
        price = i.find('span' , class_ = 'price')
        if price:
            PRICE.append(price.text.strip())
        else:
            PRICE.append(np.nan)

        #Country
        coun = i.find('span' , class_ = 'line-clamp-3 text-sm')
        country = re.findall(r'(Bourbon|Rye|Tennessee|Kentucky|Barrel|Eagle|Rum Finish|American|Irish|Very Rare|Japanese|Indian|English|Canadian|Welsh|Swedish|Formula 1|Suntory|Harmony)' , coun.text)
        if country:
            COUNTRY.append(country[0])
        else:
            COUNTRY.append(np.nan)

BRAND_VARIATION
```

#### `Replacing empty space with NAS`
- No Age Statement (NAS)
- Because some brands does not disclose age of distillery
### No Age Statement (NAS) whiskies are more common now because:

- They let distillers blend younger and older stock for consistency.

- They offer flexibility and avoid limiting perception by age alone.

```python
replace0 = lambda x : 'NAS' if x == '' else x
# replace_nan = lambda x : 'NAS' if x == np.nan else x
```

```python
df['Year'] = df['Year'].apply(replace0).head(50)
df['Year'] = df['Year'].fillna('NAS')
```

```python
#Rechecking 
df['Year'].isna().sum()
```

### `Removing Special characters and alphabet from ABV_Percent column and converting it to integer`

```python
df['ABV_Percent'] = df['ABV_Percent'].replace('\D' , '' , regex = True).astype(int)
```

#### `Replacing Pound symbol with empty space and converting it to float`

```python
df['Price_Pounds']= df['Price'].replace('[,]' , '' , regex = True).astype(float)
```

### `Creating a new feature Price_Rupees`

```python
df['Price_Rupees'] = df['Price_Pounds'] * 113.60
```

### `Creating a 'ABV_Category' Column`

```python
def abv_category(x):
    if x < 40:
        return 'Low'
    elif x > 40 and x <47:
        return 'Medium'
    else:
        return 'High'
```

```python
df['ABV_Category'] = df['ABV_Percent'].apply(abv_category)
```
